**HDBSCAN** (*Hierarchical Density-Based Spatial Clustering of Applications with Noise*)

```python
import hdbscan
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
```

```python
# Generate synthetic data
X, _ = make_moons(n_samples=500, noise=0.1, random_state=42)
```

```python
# Fit HDBSCAN model
clusterer = hdbscan.HDBSCAN(min_cluster_size=10, min_samples=5)
cluster_labels = clusterer.fit_predict(X)
```

```python
# Plotting the clusters
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels, cmap='viridis', marker='o', edgecolor='k', s=50)
plt.colorbar()
plt.title("HDBSCAN Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

Example of real dataset

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import hdbscan

# Load the dataset
df = pd.read_csv('/content/Mall_Customers.csv')
```

```python
df.head()
```

```python
# Select features for clustering
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
# Plot the dataset to see the data distribution
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', data=df)
plt.title('Customer Data Distribution')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
# Apply HDBSCAN
clusterer = hdbscan.HDBSCAN(min_cluster_size=3, min_samples=6)
labels = clusterer.fit_predict(X)
```

```python
# Add labels to the dataset for visualization
df['Cluster'] = labels
```

```python
# Plot clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', data=df, palette='viridis', legend='full')
plt.title('HDBSCAN Clustering of Customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='Cluster')
plt.show()
```
