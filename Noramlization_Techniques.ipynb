**plot a 5 x 5 matrix of histograms (continuous values) and column charts (discrete values) to visualize the distributions of each of the 25 features in the automobile dataset.**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.figure(figsize=(20, 20))
sns.set(style="white", palette="muted")
colors = plt.cm.viridis(np.linspace(0, 1, len(X.columns)))

for i, col in enumerate(X.columns):
    plt.subplot(5, 5, i + 1)
    sns.histplot(X[col], kde=True, color=colors[i], edgecolor="black")
    plt.title(col)
    plt.tight_layout()

plt.show()
```

**Min-Max Nornmalization**

```python
def min_max_normalize(series):
    return (series - series.min()) / (series.max() - series.min())

normalized_highwaympg = min_max_normalize(X['highway-mpg'])
```

```python
import matplotlib.pyplot as plt

# Assuming X['highway-mpg'] contains the original data
original_data = X['highway-mpg']
normalized_data = min_max_normalize(X['highway-mpg'])

# Plot histograms
plt.figure(figsize=(12, 5))

# Original data histogram
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st plot
plt.hist(original_data, bins=20, color='yellow', edgecolor='black')
plt.title('Original highway-mpg')
plt.xlabel('MPG')
plt.ylabel('Frequency')

# Normalized data histogram
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd plot
plt.hist(normalized_data, bins=20, color='pink', edgecolor='black')
plt.title('Normalized highway-mpg')
plt.xlabel('Normalized MPG')
plt.ylabel('Frequency')

# Show the plot
plt.tight_layout()
plt.show()
```

**Z-Score Normalization (Standardization)**

```python
def z_score_normalize(series):
    return (series - series.mean()) / series.std()

standardized_height = z_score_normalize(X['height'])
```

```python
import matplotlib.pyplot as plt

# Assuming X['highway-mpg'] contains the original data
original_data = X['height']
normalized_data = z_score_normalize(X['height'])

# Plot histograms
plt.figure(figsize=(12, 5))

# Original data histogram
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st plot
plt.hist(original_data, bins=20, color='yellow', edgecolor='black')
plt.title('Original height')
plt.xlabel('height')
plt.ylabel('Frequency')

# Normalized data histogram
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd plot
plt.hist(normalized_data, bins=20, color='pink', edgecolor='black')
plt.title('Normalized height')
plt.xlabel('Normalized Height')
plt.ylabel('Frequency')

# Show the plot
plt.tight_layout()
plt.show()
```

**L1 Normalization**

```python
def l1_normalization(series):
    return series / np.sum(np.abs(series))
```

```python
import matplotlib.pyplot as plt

# Assuming X['highway-mpg'] contains the original data
original_data = X['highway-mpg']
normalized_data = l1_normalization(X['highway-mpg'])

# Plot histograms
plt.figure(figsize=(12, 5))

# Original data histogram
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st plot
plt.hist(original_data, bins=20, color='yellow', edgecolor='black')
plt.title('Original highway-mpg')
plt.xlabel('MPG')
plt.ylabel('Frequency')

# Normalized data histogram
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd plot
plt.hist(normalized_data, bins=20, color='pink', edgecolor='black')
plt.title('Normalized highway-mpg')
plt.xlabel('Normalized MPG')
plt.ylabel('Frequency')

# Show the plot
plt.tight_layout()
plt.show()
```

**L2 Normalization**

```python
def l2_normalization(series):
    return series / np.sqrt(np.sum(np.square(series)))
```

```python
import matplotlib.pyplot as plt

# Assuming X['highway-mpg'] contains the original data
original_data = X['highway-mpg']
normalized_data = l2_normalization(X['highway-mpg'])

# Plot histograms
plt.figure(figsize=(12, 5))

# Original data histogram
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st plot
plt.hist(original_data, bins=20, color='yellow', edgecolor='black')
plt.title('Original highway-mpg')
plt.xlabel('MPG')
plt.ylabel('Frequency')

# Normalized data histogram
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd plot
plt.hist(normalized_data, bins=20, color='pink', edgecolor='black')
plt.title('Normalized highway-mpg')
plt.xlabel('Normalized MPG')
plt.ylabel('Frequency')

# Show the plot
plt.tight_layout()
plt.show()
```
